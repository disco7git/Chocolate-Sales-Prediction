```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,ConfusionMatrixDisplay,accuracy_score,precision_score,f1_score,recall_score
```

```python
iris_data=pd.read_csv(r"D:\RAZIQUE\ml\iris.csv")
iris_data
```

```python
x = iris_data.drop('virginica',axis=1)
y = iris_data['virginica']
print(x)
print(y)
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,random_state=1,test_size=0.3)
print(xtrain.shape)
print(xtest.shape)
```

```python
model = LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
y_prediction = model.predict(xtest)
y_prediction
```

```python
model.score(xtest,ytest)
```

```python
print("Accuracy:",accuracy_score(ytest,y_prediction))
```

```python
cm=confusion_matrix(ytest,y_prediction)
cm
```

```python
d=ConfusionMatrixDisplay(cm)
d.plot()
```

```python
print(classification_report(ytest, y_prediction))
```
